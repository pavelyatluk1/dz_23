### ЗАВДАННЯ 3.
Два верстати-автомати виготовляють однакові вироби, які потрапляють на загальний конвейер. Продуктивність першого автомата в два рази більша ніж продуктивність другого. Перший верстат-автомат виготовляє 60% виробів вищого гатунку, а другий - 84%.
а) Знайти ймовірність того, що навмання взятий виріб з конвейера - вищого гатунку.
б) Навмання взятий виріб з конвейера вищого гатунку. Знайти ймовірність того, що він виготовлений на першому верстаті.

### Вирішення

Події
А1 - навмання взятий виріб з конвейера - вищого гатунку
А2 - Навмання взятий виріб з конвейера вищого гатунку. Знайти ймовірність того, що він виготовлений на першому верстаті.
Продуктивність першого автомата N1 = 2/3 від загальної на конвеері.
Продуктивність другого автомата N2 = 1/3 від загальної на конвеері.


а) Ймовірність того що навмання взятий виріб вищого гатунку

    P(A1) = N1 * 0.6 + N2 * 0.84 = 2/3 * 0.6 + 1/3 * 0.84 = 0.68 або 68%
    
в) Навмання взятий виріб з конвейера вищого гатунку. Знайти ймовірність того, що він виготовлений на першому верстаті 

    P(A2) = N1 * 0.6 / P(A1) = 2/3 * 0.6 / 0.68 = 0.5882 або 58.82%
