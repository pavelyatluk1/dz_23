### ЗАВДАННЯ 2.
Для сигналізації про аварію встановлено два незалежно працюючі сигналізатори. Ймовірність того, що при аварії спрацює перший сигналізатор - 0,95, така ймовірність для другого - 0,9. Знайти ймовірність того, що при аварії спрацює:
а) хоча б один сигналізатор;
в) тільки один сигналізатор;

### Вирішення


Події:
А - спрацює перший сигналізатор - А = 0,95
В - спрацює другий сигналізатор - А = 0,9
А_ - не спрацює перший сигналізатор - А_ = 0,05
В_ - не спрацює другий сигналізатор - В_ = 0,1
С - спрацює тільки один сигналізатор

а) Для знаходження відповіді на перше питання викристовуємо ТЕОРЕМУ ДОДАВАННЯ, згідно якої вірогідність появи хоча б одної з двох сумісних подій дорівнює сумі ймовірностей цих подій без ймовірності їх спільної появи:

P(A + B) = P(A) + P(B) - P(AB) = 0.95 + 0.9 - (0.95 + 0.9) = 0.995 або 99,5%

в) Для знаходження відповіді на перше питання викристовуємо ТЕОРЕМУ ДОДАВАННЯ І МНОЖЕННЯ:

P(C) = P(A) * P(B_) + P(B) * P(A_) = 0.95 * 0.1 + 0.9 * 0.05 = 0.14 або 14%

```python
A = 0.95
B = 0.9
A_ = 0.05
B_ = 0.1
print(f"Ймовірність спрацювання хоча б одного сигналізатора - ", A + B - A * B)
print(f"Ймовірність спрацювання тільки одного сигналізатора - ", A * B_ +  A_ * B)    
    
```
